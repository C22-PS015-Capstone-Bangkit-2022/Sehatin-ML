```python
def listingxml(sourcexml):
    xmlfiles = []
    for xmlfile in os.listdir(sourcexml):
        namafile = re.search(r'(.*)\.', xmlfile)
        namafile = namafile.group(1)
        xmlfiles.append(namafile)
    return xmlfiles

def pickimage(source_image_dir, xmllist, target_image_dir):
    images = []
    for image in os.listdir(source_image_dir):
        nama_image = re.search(r'(.*)\.', image)
        nama = nama_image.group(1)
        if nama in xmllist:
            copyfile(os.path.join(source_image_dir, image), os.path.join(target_image_dir, image))
            print('copying {}'.format(image))

def convertpngtojpg(source_image_dir):
    images = []
    select_dir = source_image_dir

    for image in os.listdir(select_dir):
        ext = re.search(r'(.*)\.(.*)', image)
        if ext.group(2) != 'jpg':
            im = Image.open(os.path.join(select_dir, image))
            rgb_im = im.convert('RGB')
            rgb_im.save(os.path.join(select_dir, '{}.jpg'.format(ext.group(1))))
            
            print('converting {}'.format(image))
```

```python
def split_od(source_image, source_xml, train_dir, val_dir, split_size):
    images = os.listdir(source_image)
    xmls = os.listdir(source_xml)

    images = random.sample(images, len(images))
    train_size = int(split_size * len(images))
    val_size = len(images) - train_size

    print('remove all exists file in train and val dir')
    if os.listdir(train_dir) != []:
        for sfile in os.listdir(train_dir):
            os.remove(os.path.join(train_dir, sfile))
    if os.listdir(val_dir) != []:
        for sfile in os.listdir(val_dir):
            os.remove(os.path.join(val_dir, sfile))
    
    print('spliting images and xmls')
    for i in range(train_size):
        img = images[i]
        name = re.search(r'(.*)\.(.*)', img)
        xml = name.group(1)+'.xml'
        copyfile(os.path.join(source_image, img), os.path.join(train_dir, img))
        copyfile(os.path.join(source_xml, xml), os.path.join(train_dir, xml))

    for i in range(val_size):
        img = images[i + train_size]
        name = re.search(r'(.*)\.(.*)', img)
        xml = name.group(1)+'.xml'
        copyfile(os.path.join(source_image, img), os.path.join(val_dir, img))
        copyfile(os.path.join(source_xml, xml), os.path.join(val_dir, xml))
```

```python
source_image = os.path.join(os.getcwd(), 'selected')
source_xml = os.path.join(os.getcwd(), 'od-pascal')
train_dir = os.path.join(os.getcwd(), 'training')
val_dir = os.path.join(os.getcwd(), 'validation')
split_size = 0.9
split_od(source_image, source_xml, train_dir, val_dir, split_size)
```
